```python
import numpy as np
from scipy.io import loadmat,savemat
```

```python
base = loadmat(r'C:\Users\user\Downloads\basemodel.mat')['base_vertices'].reshape(900,1)
```

```python
model2 = loadmat(r'C:\Users\user\Downloads\model2.mat')['f'].reshape(900,1)
model3 = loadmat(r'C:\Users\user\Downloads\model3.mat')['f'].reshape(900,1)
model4 = loadmat(r'C:\Users\user\Downloads\model4.mat')['f'].reshape(900,1)
model5 = loadmat(r'C:\Users\user\Downloads\model5.mat')['f'].reshape(900,1)
model6 = loadmat(r'C:\Users\user\Downloads\model6.mat')['f'].reshape(900,1)
model7 = loadmat(r'C:\Users\user\Downloads\model7.mat')['f'].reshape(900,1)
model8 = loadmat(r'C:\Users\user\Downloads\model8.mat')['f'].reshape(900,1)
model9 = loadmat(r'C:\Users\user\Downloads\model9.mat')['f'].reshape(900,1)
model10 = loadmat(r'C:\Users\user\Downloads\model10.mat')['f'].reshape(900,1)
```

```python
#deformation = true - base
def2 = model2-base
def3 = model3-base
def4 = model4-base
def5 = model5-base
def6 = model6-base
def7 = model7-base
def8 = model8-base
def9 = model9-base
def10 = model10-base
```

```python
def_matrix = np.hstack((def2,def3,def4,def5,def6,def7,def8,def9,def10))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=9)
def_matrix_pca = pca.fit_transform(def_matrix)
def_matrix_pca.shape
```

```python
def_matrix_pca_inv = np.linalg.pinv(def_matrix_pca)
def_matrix_pca_inv.shape
```

```python
# coef_n = def_matrix_pca_inv x def_n
coef2 = np.matmul(def_matrix_pca_inv,def2)
coef3 = np.matmul(def_matrix_pca_inv,def3)
coef4 = np.matmul(def_matrix_pca_inv,def4)
coef5 = np.matmul(def_matrix_pca_inv,def5)
coef6 = np.matmul(def_matrix_pca_inv,def6)
coef7 = np.matmul(def_matrix_pca_inv,def7)
coef8 = np.matmul(def_matrix_pca_inv,def8)
coef9 = np.matmul(def_matrix_pca_inv,def9)
coef10 = np.matmul(def_matrix_pca_inv,def10)
```

```python
#def_n = def_matrix_pca x coef_n
new_def_2 = np.matmul(def_matrix_pca,coef2)
new_model_2 = base + new_def_2
new_model_2 = new_model_2.reshape(300,3)
```

```python
a = {}
a['f'] = new_model_2
savemat('test1.mat',a)
```
